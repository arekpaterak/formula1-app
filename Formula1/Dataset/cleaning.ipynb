```python
import pandas as pd
```

```python
races = pd.read_csv('kaggle/races.csv')
races = races[['raceId', 'year', 'round', 'circuitId', 'name', 'date', 'url']]

races = races[races['year'] == 2023]

races.columns = [column[0].upper() + column[1:] for column in races.columns]

races = races.reset_index(drop=True)

display(races)

races.to_csv('final/races.csv', index=False)
```

```python
circuits = pd.read_csv('kaggle/circuits.csv')
circuits = circuits.drop(columns=['circuitRef', 'lat', 'lng', 'alt'])
circuits.columns = [column[0].upper() + column[1:] for column in circuits.columns]

circuits_in_2023 = races['CircuitId'].unique()
circuits = circuits[circuits['CircuitId'].isin(circuits_in_2023)]

circuits.rename(columns={'Location': 'City'}, inplace=True)

display(circuits)

circuits.to_csv('final/circuits.csv', index=False)
```

```python
teams = pd.read_csv('kaggle/constructors.csv')
teams = teams.drop(columns=['constructorRef'])

teams = teams.rename(columns={'constructorId': 'teamId'})
teams.columns = [column[0].upper() + column[1:] for column in teams.columns]
```

```python
drivers = pd.read_csv('kaggle/drivers.csv')
drivers.columns = [column[0].upper() + column[1:] for column in drivers.columns]
drivers = drivers.drop(columns=['DriverRef'])
drivers = drivers.rename(columns={'Forename': 'FirstName', 'Surname': 'LastName'})
```

```python
race_results = pd.read_csv('github/Formula1_2023season_raceResults.csv')

race_results.columns = race_results.columns.str.replace(' ', '')
race_results = pd.concat([race_results, race_results['Driver'].str.split(' ', n=1, expand=True).rename(columns={0: 'FirstName', 1: 'LastName'})], axis=1)
race_results['LastName'] = race_results['LastName'].str.replace('De ', 'de ')
race_results['LastName'] = race_results['LastName'].str.replace('Perez', 'PÃ©rez')
race_results['LastName'] = race_results['LastName'].str.replace('Hulkenberg', 'HÃ¼lkenberg')

track_to_circuit_id = {
    'Bahrain': 3,
    'Saudi Arabia': 77,
    'Australia': 1,
    'Azerbaijan': 73,
    'Miami': 79,
}

race_results['CircuitId'] = race_results['Track'].apply(lambda x: track_to_circuit_id[x])

team_to_team_id = {
    'Red Bull Racing Honda RBPT': 9, 
    'Aston Martin Aramco Mercedes': 117,
    'Ferrari': 6, 
    'Mercedes': 131, 
    'Alfa Romeo Ferrari': 51, 
    'Alpine Renault': 214,
    'Williams Mercedes': 3, 
    'AlphaTauri Honda RBPT': 213, 
    'Haas Ferrari': 210,
    'McLaren Mercedes': 1
}

race_results['TeamId'] = race_results['Team'].apply(lambda x: team_to_team_id[x])

race_results = race_results.merge(drivers, on=['LastName', 'FirstName'], how='left')[[*race_results.columns.tolist(), 'DriverId']]

race_results = race_results.merge(races, on='CircuitId', how='left')[[*race_results.columns.tolist(), 'RaceId']]

race_results = race_results.drop(columns=['Track', 'Driver', 'Team', 'FirstName', 'LastName', 'No', 'CircuitId'])

columns = race_results.columns.to_list()
columns = list(reversed(columns[-3:])) + columns[:-3]
race_results = race_results[columns]

race_results['Sprint'] = 0
columns = race_results.columns.to_list()
columns = columns[:-3] + columns[-1:] + columns[-3:-1]
race_results = race_results[columns]

display(race_results)

race_results.to_csv('final/race_results.csv', index=False)
```
